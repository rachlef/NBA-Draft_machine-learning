```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor

# Path of file to read
data_path = "csv/CollegeBasketballPlayers2009-2021.csv"

# Reading the data from the csv file
total_ball_data = pd.read_csv(data_path, low_memory = False)

# Only include rows in which a player was drafted, and before 2021 (2021 will be used as test data)
ball_data_1 = total_ball_data.loc[total_ball_data['pick'] >= 1]
ball_data = ball_data_1.loc[ball_data_1['year'] < 2021]
```

```python
# Shows a preview of data
ball_data.head()
```

```python
# Define 'pick' as the target variable, which we'll refer to as y
y = ball_data.pick

# Create feature matrix X, using a specific set of basketball statistics
features = ['Min_per', 'ftr', 'pts', 'ORB_per', 'DRB_per', 'blk_per', 'stl_per', 'dporpag', 'porpag', 'AST_per', 'TO_per', 'FT_per']
X = ball_data[features]

# Partition the data into training and validation sets
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=1)
```

```python
# Establish a Decision Tree Regressor as our model, with specific parameters
basketball_model = DecisionTreeRegressor(max_leaf_nodes=100, splitter='random', random_state=1, max_features='auto')

# Train the model using our training data
basketball_model.fit(train_X, train_y)

# Use the trained model to make predictions on the validation data and compute the mean absolute error
predicted_values = basketball_model.predict(val_X)
mean_abs_error = mean_absolute_error(predicted_values, val_y)
print("Mean Absolute Error on validation data: {:,.5f}".format(mean_abs_error))
```

```python
# Select data from the year 2021 for testing
data_2021 = ball_data_1.loc[ball_data_1['year'] == 2021]

# Define features
features = ['Min_per', 'ftr', 'pts', 'ORB_per', 'DRB_per', 'blk_per', 'stl_per', 'dporpag', 'porpag', 'AST_per', 'TO_per', 'FT_per']

# Generate test features for 2021 data
test_features_2021 = data_2021[features]

# Generate predictions using the test data
predictions_2021 = basketball_model.predict(test_features_2021)

# Calculate the mean absolute error for the predictions
mae_2021 = mean_absolute_error(predictions_2021, data_2021.pick)

# Prepare predictions in the required format and save them to a CSV file
prediction_output = pd.DataFrame({'Player Name': data_2021.player_name, 'Predicted Pick': predictions_2021, 'Actual Pick': data_2021.pick})
prediction_output.to_csv('predictions2021.csv', index=False)

print("Mean Absolute Error for 2021 Test Data: {:,.5f}".format(mae_2021))
```

```python
from xgboost import XGBRegressor

# Instantiate the XGBoost regressor model
xgb_model = XGBRegressor(n_estimators=1200, learning_rate=0.106, max_depth=10, reg_lambda=1.1, tree_method='approx', random_state=1)

# Train the model with the training data
xgb_model.fit(train_X, train_y)

# Generate predictions on the validation data
xgb_predictions = xgb_model.predict(val_X)

# Calculate the mean absolute error of the predictions
mae_xgb = mean_absolute_error(xgb_predictions, val_y)

print("Mean Absolute Error for XGBoost Validation Data: {:,.5f}".format(mae_xgb))
```

```python
# Extract data for the year 2021
basketball_data_2021 = ball_data_1.loc[ball_data_1['year'] == 2021]

# Define the features you're interested in
feature_list = ['Min_per', 'ftr', 'pts', 'ORB_per', 'DRB_per', 'blk_per', 'stl_per', 'dporpag', 'porpag', 'AST_per', 'TO_per', 'FT_per']

# Create your 2021 test data
test_data_2021 = basketball_data_2021[feature_list]

# Use the XGBoost model to generate predictions for the 2021 data
xgb_predictions_2021 = xgb_model.predict(test_data_2021)

# Compute the mean absolute error for the XGBoost predictions
mae_xgb_2021 = mean_absolute_error(xgb_predictions_2021, basketball_data_2021.pick)

print("Mean Absolute Error for XGBoost on 2021 Test Data: {:,.5f}".format(mae_xgb_2021))
```

```python
# Prepare the predictions for saving
predictions_df = pd.DataFrame({'Player Name': basketball_data_2021.player_name, 'Predicted Pick': xgb_predictions_2021, 'True Pick': basketball_data_2021.pick})

# Write the predictions to a CSV file
predictions_df.to_csv('xgb_predictions_2021.csv', index=False)
```

```python
# Random Forest Regressor: This model is an ensemble learning method that operates by constructing a multitude of decision trees at training time and outputting the mean prediction of the individual trees.

from sklearn.ensemble import RandomForestRegressor

# Instantiate the Random Forest Regressor
rf_model = RandomForestRegressor(n_estimators=100, random_state=1)

# Train the model
rf_model.fit(train_X, train_y)

# Make predictions using the validation data
rf_preds = rf_model.predict(val_X)

# Calculate the Mean Absolute Error (MAE)
rf_val_mae = mean_absolute_error(rf_preds, val_y)

print("Validation MAE for Random Forest: {:,.5f}".format(rf_val_mae))
```

```python
# Feature Importance: Many models can output the importance of each feature in making predictions. Here's how you might do it for the Random Forest model:
import matplotlib.pyplot as plt

importances = rf_model.feature_importances_
indices = np.argsort(importances)

plt.figure(figsize=(10,10))
plt.title('Feature Importances')
plt.barh(range(len(indices)), importances[indices], color='b', align='center')
plt.yticks(range(len(indices)), [features[i] for i in indices])
plt.xlabel('Relative Importance')
plt.show()
```

```python
#Actual vs. Predicted Plots: These plots can show how close your model's predictions are to the actual values.
plt.figure(figsize=(10,10))
plt.scatter(val_y, rf_preds)
plt.plot([min(val_y), max(val_y)], [min(val_y), max(val_y)], color='red')
plt.title('Actual vs Predicted')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.show()
```

```python
# Residual Plots: These plots can show the difference between actual and predicted values for each data point. Ideally, residuals should be randomly and evenly scattered around zero.
residuals = val_y - rf_preds

plt.figure(figsize=(10,10))
plt.scatter(rf_preds, residuals)
plt.title('Residual Plot')
plt.xlabel('Prediction')
plt.ylabel('Residuals')
plt.show()
```

```python
# Support Vector Regression (SVR): This model is an extension of support vector machines (SVM) for regression problems.

from sklearn.svm import SVR

# Instantiate the Support Vector Regression
svr_model = SVR(kernel='linear')

# Train the model
svr_model.fit(train_X, train_y)

# Make predictions using the validation data
svr_preds = svr_model.predict(val_X)

# Calculate the Mean Absolute Error (MAE)
svr_val_mae = mean_absolute_error(svr_preds, val_y)

print("Validation MAE for SVR: {:,.5f}".format(svr_val_mae))
```

```python
import matplotlib.pyplot as plt

# Actual vs Predicted Plot
plt.figure(figsize=(10,10))
plt.scatter(val_y, svr_preds)
plt.plot([min(val_y), max(val_y)], [min(val_y), max(val_y)], color='red')
plt.title('Actual vs Predicted - SVR')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.show()

# Residual Plot
residuals = val_y - svr_preds

plt.figure(figsize=(10,10))
plt.scatter(svr_preds, residuals)
plt.title('Residual Plot - SVR')
plt.xlabel('Prediction')
plt.ylabel('Residuals')
plt.show()
```

```python
# Gradient Boosting Regressor: This model produces a prediction model in the form of an ensemble of weak prediction models, typically decision trees.

from sklearn.ensemble import GradientBoostingRegressor

# Instantiate the Gradient Boosting Regressor
gbr_model = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=1, random_state=1)

# Train the model
gbr_model.fit(train_X, train_y)

# Make predictions using the validation data
gbr_preds = gbr_model.predict(val_X)

# Calculate the Mean Absolute Error (MAE)
gbr_val_mae = mean_absolute_error(gbr_preds, val_y)

print("Validation MAE for Gradient Boosting: {:,.5f}".format(gbr_val_mae))
```

```python
import matplotlib.pyplot as plt

# Actual vs Predicted Plot
plt.figure(figsize=(10,10))
plt.scatter(val_y, gbr_preds)
plt.plot([min(val_y), max(val_y)], [min(val_y), max(val_y)], color='red')
plt.title('Actual vs Predicted - Gradient Boosting')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.show()

# Residual Plot
residuals = val_y - gbr_preds

plt.figure(figsize=(10,10))
plt.scatter(gbr_preds, residuals)
plt.title('Residual Plot - Gradient Boosting')
plt.xlabel('Prediction')
plt.ylabel('Residuals')
plt.show()

# Feature Importance
importances = gbr_model.feature_importances_
indices = np.argsort(importances)

plt.figure(figsize=(10,10))
plt.title('Feature Importances - Gradient Boosting')
plt.barh(range(len(indices)), importances[indices], color='b', align='center')
plt.yticks(range(len(indices)), [features[i] for i in indices])
plt.xlabel('Relative Importance')
plt.show()
```
